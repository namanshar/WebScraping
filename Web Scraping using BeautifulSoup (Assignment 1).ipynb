# Question 1: Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
#Webpage source code
page.content
```

```python
#Parsing the sorce code by Beautiful soup
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
#Extracting the tag for all the header tage of webpage
h_t = soup.find_all('span', class_='mw-headline' )
h_t
```

```python
#Displaying only text
for i in h_t:
 print(i.text)
```

```python
#Storing all the header tags in a list
htag = []
for i in h_t:
 htag.append(i.get_text())
```

```python
htag
```

```python
#Storing into a dataframe
import pandas as pd
h_list=pd.DataFrame({})
h_list['Header']=htag
```

# Question 2: Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
#TOP 1-100 movies
page1 = requests.get('https://www.imdb.com/chart/top/')
page1.content
```

```python
soup1 = BeautifulSoup(page1.content,'html.parser')
soup1
```

```python
#Empty lists for storing name, rank and year.
name= []
rank = []
year = []
```

```python
rank = soup1.find_all('span', class_='lister-item-index unbold text-primary' )
rank
```

```python
data = soup1.find_all(['tr'])
```

```python
data
```

```python
for i in range(0,len(data)):
    if(i>0 and i<=100):        
        name_=data[i].find('td',class_="titleColumn")
        name.append(name_.find('a').get_text())
        
        rank_=data[i].find('td',class_="ratingColumn imdbRating")
        rank_=rank_.find('strong').get_text()
        rank.append(rank_)
        
        year_=name_.find('span').get_text().replace('(','')
        year_=year_.replace(')','')
        year.append(year_)
        
```

```python
#Creating a data frame from above data
df=pd.DataFrame({})
df['Name']=name
df['Year of release']=year
df['IMDB rank']=rank
```

```python
df
```

```python
#Saving dataframe into csv file
df.to_csv('TopIMDB.csv')
```
